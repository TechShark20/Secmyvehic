```python
# @title Error while recognising
#@markdown error or system failure


perfectrun=0
systemfail=0
time="time"
error="error"
for i in range(1,6):
  perfectrun+=len(globals()[time+str(i)])
  systemfail+=globals()[error+str(i)]
print("*****************************************************************")
print(" the error rate in while recongising or aligning the face was {0:.4f}".format(perfectrun/(perfectrun+systemfail)))
print("*****************************************************************")

cat=['perfect','failure']
plt.figure(figsize=(6, 3))
val=[((perfectrun)/(perfectrun+systemfail))*100,-((systemfail)/(perfectrun+systemfail))*100]
plt.barh(cat,val,color=['green', 'red'])
plt.xlim(-100,100)
plt.xlabel('Percentage')

plt.grid(True)
plt.title('Face alignment analysis ')
plt.savefig('Facealignmentana.png', dpi=300)
plt.show()

```

```python
#@title CONFUSION MATRIX

counttp="counttp"
counttn="counttn"
countfp="countfp"
countfn="countfn"
TP=0  #combined true positive
TN=0 #combined true negative
FP=0 # combined false positive
FN=0 # combined false negative
for i in range(1,6):
  TP+=globals()[counttp+str(i)]
  TN+=globals()[counttn+str(i)]
  FP+=globals()[countfp+str(i)]
  FN+=globals()[countfn+str(i)]



Accuracy=(TP+TN)/(TP+TN+FP+FN)
print("ACcuracy for the system is ::",Accuracy*100)
print(" ")
cnfmatrix=np.array([[TP,FP],[FN,TN]])
precision =TP/(TP+FP)
print("Precision for the system is ::",precision )
print(" ")
Recall =TP/(TP+FN)
print("Recall for the system is ::",Recall)
print(" ")
F1score=2/((1/precision)+(1/Recall))
print("F1 score for the system is  ",F1score)
```

```python

#@title  plotting the confusion matrix
plt.title("True Label",fontsize=10)

sns.heatmap(cnfmatrix,fmt='d',annot=True, cmap='Blues', cbar=False, yticklabels=['Predicted 1', 'Predicted 0'], xticklabels=['True 1', 'True 0'])
plt.tick_params(axis='x', which='both', bottom=False, top=True, labelbottom=False, labeltop=True)
plt.ylabel('Predicted Label')
plt.xlabel('Analysis of overall performance ',fontsize=12,labelpad=20)
plt.tight_layout()
plt.savefig('confusion.png', dpi=300)
```
